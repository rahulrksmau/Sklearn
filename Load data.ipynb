```python
dataset = [train,test]
for data in dataset:
    data.drop('PassengerId', axis=1, inplace=True)
    
for data in dataset:
    data['Name_len'] = data['Name'].apply(len)

for data in dataset:
    data['FamilySize'] = data['SibSp'] + data['Parch']+1

for data in dataset:
    data['IsAlone'] = 0
    data.loc[data['FamilySize'] == 1, 'IsAlone'] = 1
    
```

```python
for data in dataset:
    data['Embarked'] = data['Embarked'].fillna('S')
    
for date in dataset:
    data['Fare'] = data['Fare'].fillna(data['Fare'].median())

train['Fare_category'] = pd.qcut(train['Fare'],4)
test['Fare_category'] = pd.qcut(test['Fare'],4)
for data in dataset:
    avg = data['Age'].mean()
    std = data['Age'].median()
    size= data['Age'].isnull().sum()
    data_to_fill = np.random.randint(avg-std, avg+std, size=size)
    data['Age'][np.isnan(data['Age'])] = data_to_fill
    data['Age'] = data['Age'].astype(int)
train['Age_category'] = pd.qcut(train['Age'], 5)
test['Age_category'] = pd.qcut(test['Age'], 5)
```

```python
train['Age_category'].head()
```

```python
import re
def get_title(name):
    title_search = re.search(' ([A-Za-z]+)\.', name)
    # If the title exists, extract and return it.
    if title_search:
        return title_search.group(1)
    return ""
male_col = ['Sir','Col','Master']
female_col = ['Lady','Mme','Mlle','Ms']
other_col = ['Dr', 'Rev','Major','Don','Capt','Jonkheer','Dona','Countess']
for data in dataset:
    data['Title'] = data['Title'].replace(male_col,'Mr')
    data['Title'] = data['Title'].replace(female_col,'other')
    data['Title'] = data['Title'].replace(other_col,'other')    
```

```python
train['Has_Cabin'] = train["Cabin"].apply(lambda x: 0 if type(x) == float else 1)
test['Has_Cabin'] = test["Cabin"].apply(lambda x: 0 if type(x) == float else 1)
```

```python
from sklearn import preprocessing
features = ['Sex','Embarked','Fare_category','Age_category','Title']
df_combined = pd.concat([train[features],test[features]])
for feature in features:
    le = preprocessing.LabelEncoder()
    le = le.fit(df_combined[feature])
    train[feature] = le.transform(train[feature])
    test[feature]  = le.transform(test[feature])
```
