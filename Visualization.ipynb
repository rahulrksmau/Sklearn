```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import pandas as pd
train = pd.read_csv('modified_train.csv')
test = pd.read_csv('modified_test.csv')
```

```python
g = sns.pairplot(train[[u'Survived', u'Pclass', u'Sex', u'Embarked',u'FamilySize',u'IsAlone',u'Fare_category',u'Age_category', u'Has_Cabin' ,u'Title']], hue='Survived', palette = 'seismic',size=1.2,diag_kind = 'kde',diag_kws=dict(shade=True),plot_kws=dict(s=10) )
g.set(xticklabels=[])
```
